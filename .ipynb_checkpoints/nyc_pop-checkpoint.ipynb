```python
# Importing dependencies
import json
import pandas as pd
import datetime
from pprint import pprint
```

```python
nyc_json = json.load(open("nyc_pop.json"))
```

```python
pprint(nyc_json)
```

```python
nyc_pop_df = pd.DataFrame(nyc_json["data"], columns=[
  # Dropping
  "drop1","drop2","drop3","drop4","drop5","drop6","drop7","drop8","total_population",
  # Keeping
  "borough",
  # Dropping
  "1950","1950%","1960","1960%","1970","1970%","1980","1980%","1990","1990%",
  "2000","2000%","2010","2010%",
  # Keeping
  "2020","2020%",
  # Dropping
  "2030","2030%","2040","2040%"
])
drop_columns = [
  "total_population","drop1","drop2","drop3","drop4","drop5","drop6","drop7","drop8","1950","1950%","1960",
  "1960%","1970","1970%","1980","1980%","1990","1990%","2000","2000%","2010","2010%",
  "2030","2030%","2040","2040%"
]
nyc_pop_df = nyc_pop_df.drop(columns=drop_columns)
nyc_pop_df = nyc_pop_df.rename(columns={"2020":"total_population", "2020%":"pop_portion"})
nyc_pop_df['borough'] = nyc_pop_df['borough'].apply(lambda x: str.strip(x))

nyc_pop_df.to_csv("2020_nyc_pop_df.csv")
```
